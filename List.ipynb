```python
# indexing
L = [1,2,3,4,5,6,7,8,9]
print(L[0])
print(L[-1])
```

# List Function

```python
# min/len/max/sorted

L = [5,7,2,0,7,3]
print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L,reverse=True))
```

```python
# reverse
L = [5,7,2,0,7,3]
# premanently in reverse
L.reverse()
print(L)
```

```python
# sort ( vs sorted)
L = [86,54,23,64,23]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

# List Comprehension
 * List comprehension provides a concise way of creating lists
 * newlist = [expression for item in iterable if condition == True]

```python
# add 1 to 10 in number list
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
[s*i for i in v]
```

```python
# add squares
L = [1,2,3,4,5]
[i**2 for i in L]
```

```python
# print all number div by 5 in the range of 50

[i for i in range(1,51) if i%5==0]
```

```python
# Nested if with List Comprehension
basket = ['apple','gau','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
# [fruit for fruit in my_fruits if fruit in basket if fruit[0]=='a']
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# cartesian product
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```
